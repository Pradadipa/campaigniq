```python
import matplotlib.pyplot as plt

# Example data (e.g., monthly website traffic)
months = list(range(1, 13))
traffic = [12000, 13000, 15000, 18000, 21000, 25000, 
           15000, 14000, 14500, 16000, 17000, 19000]

plt.figure(figsize=(10, 6))

# Plot the traffic trend
plt.plot(months, traffic, marker='o', linewidth=2)

# Add title and labels
plt.title("Website Organic Traffic in 2024")
plt.xlabel("Month")
plt.ylabel("Traffic")

# Add context annotation (your code)
plt.annotate(
    'Organic traffic declined 34% due to competitor SEO push',
    xy=(7, 15000),          # Point you're annotating
    xytext=(8, 20000),      # Where the text appears
    arrowprops=dict(arrowstyle='->')
)

# Improve visualization
plt.grid(True, linestyle='--', alpha=0.4)
plt.xticks(months)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
months = list(range(1, 13))
traffic = [12000, 13000, 15000, 18000, 21000, 25000, 
           15000, 14000, 14500, 16000, 17000, 19000]

target_value = 18000  # Benchmark line

plt.figure(figsize=(10, 6))

# Plot the traffic trend
plt.plot(months, traffic, marker='o', linewidth=2, label="Organic Traffic")

# Add benchmark line
plt.axhline(y=target_value, color='gray', linestyle='--', 
            label=f'Target: {target_value}')

# Add annotation
plt.annotate(
    'Organic traffic declined 34% due to competitor SEO push',
    xy=(7, 15000),           # Point you're explaining
    xytext=(8.2, 20000),     # Text placement
    arrowprops=dict(arrowstyle='->')
)

# Labels and title
plt.title("Website Organic Traffic in 2024")
plt.xlabel("Month")
plt.ylabel("Traffic")
plt.xticks(months)
plt.grid(True, linestyle='--', alpha=0.3)

# Show legend for benchmark line
plt.legend()

plt.show()
```
